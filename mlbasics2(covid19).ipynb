```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
#keeping only required columns
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
df.head()
```

```python
df.columns=['date','state','cured','deaths','confirmed']
df.head()
```

```python
df.tail()
```

```python
today=df[df.date=='2021-06-08']
today.head()
```

```python
today.tail()
```

```python
max_confirm=today.sort_values(by="confirmed",ascending=False)
max_confirm
```

```python
top_states=max_confirm[0:5]
top_states
```

```python
#making barplot for states with top confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="confirmed",data=top_states,hue="state")
plt.show()
```

```python
max_death=today.sort_values(by="deaths",ascending=False)
top_deaths=max_death[0:5]
top_deaths
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="deaths",data=top_deaths,hue="state")
plt.show()
```

```python
max_cured=today.sort_values(by="cured",ascending=False)
top_cured=max_cured[0:5]
top_cured
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="cured",data=top_cured,hue="state")
plt.show()
```
